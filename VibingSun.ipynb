## More about stellar pulsations

Stars whose luminosity varies periodically have been known for centuries. However,
only within the last hundred years has it been definitely established that in many cases
these variations are due to intrinsic pulsations of the stars themselves


The main emphasis in the early studies was on understanding the causes of the pulsations, particularly the concentration of pulsating stars in the instability strip


 In many instances more than one mode of oscillation is excited simultaneously in a star; these modes
may include both radial overtones, in addition to the fundamental, and nonradial modes,
where the motion does not preserve spherical symmetry

**This development is extremely important for attempts to use pulsations to learn about the properties of stars: each observed

star, and hence the amount of information about the star grows with the number of modes

## Acoustic eigenmodes 
Now that we know that our model works well we can push a bit further.\
Here we want to obtain the solar acoustic eigenmodes. This one is where you need to have some trust in me. I am going to get you some understanding, whatever it takes.\

**Some Technical Comments on the subject matter**
First for the pleasure of physics/math studentd; The matematics of the model is not that simple. The equations which give birth to the model are a set of partial differential equations (such as the continuity equation) with specific boundary conditions suited for the 3-D Sun object.\
These equations allow us to fit all types of modes possible:
- Gravity Modes – restoring force is buoyancy – internal gravity waves
- Pressure Modes – restoring force is pressure
- Fundamental Modes – restoring force is buoyancy modified by density interface – surface gravity waves
Our interest here relies on the pressure modes and these are the one we are reffering to as acoustic modes. 

**What are acoustic eigenmodes?**
Here for everyone. One way or another you actually know what an eigenmode, also called *eigenfrequency*, is.\
For example you should be familiar with the effect of acoustic resonance, else you can watch [this](https://www.youtube.com/watch?v=zopGLYB_MNw). In the video we see a speaker substituing the more famous opera singer which breaks the crystal glass with her voice. The glass will break once the voice of the singer/speaker matches the  *eigenfrequency* (also called natural frequency or resonating frequency in this context) intrinsic to the glass. Essentially this frequency tell us the preferred way for the glass to bounce back and forward when hit.

Similar illustrations can be done via instruments, for example thinking about the preferred ways of guitar strings to bounce back and forward once plucked (look it up!). In our case, we are talking about the ways in which the Sun prefers to bounce back and forward.

**Radial and Non-Radial Stellar Oscillations**
Regarding the preferred way for the Sun to bounce back and forward, the kind of bouncing which we would naturally picture in our head would be the simple radial oscillation, where all the Surfaces shrink and expands symmetrically.\
About this, from the Danish lecture notes:\
*This assumption is successful in many cases; however, radial oscillations are only a few among the many possible
oscillations of a star, and the possible presence of nonradial modes must be kept in mind in
analyses of oscillation observations*

Again:

*A nonradial mode is characterized by three wavenumbers: the degree l, and azimuthal order m which determine the behaviour of the mode overthe surface of the star and the radial order n which reflects the properties in the radial direction.* **In general the frequencies ω_nlm of stellar oscillations
depend on all three wave numbers. It is convenient, however, to separate the frequency
into the multiplet frequency ωnl**, *obtained as a suitable average over azimuthal order m and
corresponding to the spherically symmetric structure of the star*, **and the frequency splitting
δωnlm = ωnlm − ωnl**

We see from this that in order to obtain useful data for discovering properties of the sun interior we can focus on eigenmodes related to the quantum numbers n (n_p) and l.

# The Golf Observations

## Introduction to the observations
Now that we know that our model is doing good, we can actually process some experimental data and test it with it.\
The dataset chosen for this comparison comes from the GOLF experiment as mentioned above. Essentially the datasets comprises the integrated recordings of multiple observatories in different part of Earth, in order to have a continuous long measurement of the Sun surface.\
Observations of the velocity oscillations are carried out by measuring the Doppler shift of spectral lines; hence such observations are only sensitive to the line-of-sight component of velocity.\ 
What interest us is to measure the velocity dispersions around the radial velocity of the Sun, which are caused by the Sun oscillations. 

Remember that the radial velocity of a star or other luminous distant objects can be measured accurately by taking a high-resolution spectrum and comparing the measured wavelengths of known spectral lines to wavelengths from laboratory measurements.

Just to emphasize, the **dataset contains incredible measurements** containing  790560 datapoints taken 20 seconds apart from each other, consisting of a total of close to 10 years of continous measurements.\
More than this the datasets contain samples from two photomultiplicators, and each of these are elaborated by two different teams performing different calibrations. 

All four final results are displayed in what follows
![golf](dataset.png "GOLF Data")

Cal1 represents Method 1 below, Cal2 represents Method 2 below


        o Method 1 (Paris)
          Calibration based on the magnetic modulation of the signal
          The amplitude of the intensity variation induced by a Doppler
          shift of the sodium lines, is proportional to the slopes of the
          wings around the operating points. Such local slopes can be
          estimated thanks to an instrumental magnetic modulation. The
          velocity can be inferred knowing the amplitude of the magnetic
          modulation.
          Data from Photo-Multiplier 1 and Photo-Multiplier 2.

        o Method 2 (Los Angles)
          Calibration based on the modeling of the photometric response of
          the instrument.
          The values of the measured intensities are parameterised as well
          as possible, and the system is eventually solved as a function of
          the orbital velocity.
          Data from Photo-Multiplier 1 and Photo-Multiplier 2.


After visualizing the dataset we notice a difference with a better quality in terms of noise and steady trend for method 1, hence this was chosen for further comparisons.
